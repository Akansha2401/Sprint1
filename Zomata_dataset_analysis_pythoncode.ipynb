# **Removing "NEW" , "-" and "/5" from Rate Column**

```python
# Define a function to handle and clean the 'rate' values from the dataset.
def handlerate(value):
    # Check if the value is 'NEW' or '-', which indicates a non-numeric rate, and return NaN for those cases.
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        # Split the string by '/' to handle cases with multiple ratings and select the first part.
        value = str(value).split('/')
        value = value[0]  # Get the first rating value from the split.
        return float(value)  # Convert the extracted rating to a float for numerical analysis.
    
# Apply the handlerate function to the 'rate' column to clean the data.
df['rate'] = df['rate'].apply(handlerate)

# Display the first five entries of the cleaned 'rate' column.
df['rate'].head()
```

**Insight :**

This function effectively cleans the 'rate' column by converting invalid entries to NaN and extracting the first numeric rating for further analysis.

# **Filling Null Values in Rate Column with Mean**

```python
# Fill any NaN values in the 'rate' column with the mean rating to maintain data continuity.
df['rate'].fillna(df['rate'].mean(), inplace = True)

# Check the number of missing values in the 'rate' column after filling NaNs.
df['rate'].isnull().sum()

# Display the DataFrame's information, including the data types and the number of non-null entries for each column.
df.info()
```

**Insight :**

The process ensures that all missing values in the 'rate' column are addressed, allowing for a complete dataset for further analysis.

# **Removing , from Cost2Plates Column**

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)
    
df['Cost2plates'] = df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].unique()
```

# **Cleaning Rest Type Column**

```python
# Count the occurrences of each unique restaurant type in the 'rest_type' column.
rest_types = df['rest_type'].value_counts(ascending=False)

# Display the counts of restaurant types.
rest_types
```

**Insight :**

The rest_types output reveals the distribution of restaurant types, highlighting the most popular categories within the dataset.

```python
# Filter the restaurant types to include only those with fewer than 1000 occurrences
rest_types_lessthan1000 = rest_types[rest_types < 1000]  
# Display the restaurant types that are less common in the dataset
rest_types_lessthan1000  
```

**Insight :**

The rest_types_lessthan1000 variable shows the less prevalent restaurant types, indicating that the majority of the dataset is dominated by a few popular categories.

# **Making Rest Types less than 1000 in frequency as others**

```python
# Define a function to categorize restaurant types with fewer than 1000 occurrences as 'others'
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'  # Assign 'others' if the restaurant type is less common
    else:
        return value  # Keep the original type for more common restaurant types

# Apply the function to the 'rest_type' column to classify types
df['rest_type'] = df['rest_type'].apply(handle_rest_type)  
# Count the occurrences of each restaurant type after categorization
df['rest_type'].value_counts()  
```

**Insight :**

The transformation groups less common restaurant types into an "others" category, simplifying the analysis by focusing on the predominant types.

# **Cleaning Location Column**

```python
# Count the occurrences of each location in the dataset
location = df['location'].value_counts(ascending=False)  

# Filter locations that occur less than 300 times
location_lessthan300 = location[location < 300]  

# Define a function to categorize locations with fewer than 300 occurrences as 'others'
def handle_location(value):
    if(value in location_lessthan300):  
        return 'others'  # Assign 'others' if the location is less common
    else:
        return value  # Keep the original location for more common entries

# Apply the function to the 'location' column to classify locations
df['location'] = df['location'].apply(handle_location)  
# Count the occurrences of each location after categorization
df['location'].value_counts()  
```

**Insight:**

This categorization consolidates less frequent locations into an "others" category, enabling a clearer focus on the most significant locations in the dataset.

# **Cleaning Cuisines Column**

```python
# Count the occurrences of each cuisine type in the dataset
cuisines = df['cuisines'].value_counts(ascending=False)  

# Filter cuisines that occur less than 100 times
cuisines_lessthan100 = cuisines[cuisines < 100]  

# Define a function to categorize cuisines with fewer than 100 occurrences as 'others'
def handle_cuisines(value):
    if(value in cuisines_lessthan100):  
        return 'others'  # Assign 'others' if the cuisine is less common
    else:
        return value  # Keep the original cuisine for more common entries

# Apply the function to the 'cuisines' column to classify cuisines
df['cuisines'] = df['cuisines'].apply(handle_cuisines)  
# Count the occurrences of each cuisine after categorization
df['cuisines'].value_counts()  
```

**Insight :**
This categorization consolidates less common cuisines into an "others" category, simplifying the analysis and highlighting the more popular cuisine types in the dataset.

# **Data Visualization**

# **Chart 1(Visualizing Online Order)**

```python
# Group the data by 'online_order' and count the occurrences
online_order_counts = df['online_order'].value_counts()

# Set the figure size to 6x6
plt.figure(figsize=(6,6))

# Create a brighter color palette for the pie chart
colors = sns.color_palette('pastel', len(online_order_counts))

# Plot the pie chart with brighter colors
plt.pie(online_order_counts, labels=online_order_counts.index, autopct='%1.1f%%', colors=colors, startangle=90)

# Set the title for the pie chart
plt.title('Online Order Distribution')

# Display the pie chart
plt.show()
```

**Insight :**

The chart shows that 59.2% of customers prefer online orders, indicating a strong demand for online services. However, 40.8% still choose offline options, emphasizing the need for maintaining both channels.

# **Chart 6(No. of Votes, Location Wise)**
